```python
import os
import pip
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in 2014 Childrens recode
df = pd.read_sas('./../shared_data_read_only/Data/DHS/GH_2014_DHS/GHKR72SD/GHKR72FL.SAS7BDAT')
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
columns_labels = ['Region', 'Highest educational level', 'Source of drinking water', 'Religion', 'Education in single years', 'Number of children under 5', 'Literacy', 'Wealth index combined', 'Medical help money needed', 'Current marital status', 'Respondent currently working']
columns_list = ['V024', 'V106', 'V113', 'V130', 'V133', 'V137', 'V155', 'V190', 'V467C', 'V501', 'V714']
df_selected = df[columns_list]
# cov_matrix = df_selected.cov()
cov_matrix = df_selected.corr()
cax = ax.imshow(cov_matrix, cmap='coolwarm', interpolation='none', vmin=-1, vmax=1)
fig.colorbar(cax, ax=ax)

for (i, j), val in np.ndenumerate(cov_matrix):
    ax.text(j, i, f'{val:.2f}', ha='center', va='center', color='black')

# plt.xticks(np.arange(len(columns_list)), columns_labels, rotation=90)
ax.set_xticks(np.arange(len(columns_list)), columns_list, rotation=90)
ax.set_yticks(np.arange(len(columns_list)), columns_labels)
# plt.yticks(np.arange(len(columns_list)), columns_list)
plt.title('2014 Childrens recode')
plt.tight_layout()
plt.savefig("2014 Childrens recode.png", dpi=450)
plt.show()
```
